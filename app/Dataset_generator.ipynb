### Creating image dataset for Rotated images
Just provide path of the folders where images are present

```python
path_1 = './scanned_id/'
for i in os.listdir(path_1):
    for j in range(0,360,90):
        img = cv2.imread(os.path.join(path_1,i))
        if j == 0:
            rotated = img
            cv2.imwrite(os.path.join('./',(str(j)+'/'+str(i)+'_'+str(j)+'.jpg')),rotated)
        if j ==90:
            rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
            cv2.imwrite(os.path.join('./',(str(j)+'/'+str(i)+'_'+str(j)+'.jpg')),rotated)
        if j ==180:
            rotated = cv2.rotate(img, cv2.ROTATE_180)
            cv2.imwrite(os.path.join('./',(str(j)+'/'+str(i)+'_'+str(j)+'.jpg')),rotated)        
        if j ==270:
            rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
            cv2.imwrite(os.path.join('./',(str(j)+'/'+str(i)+'_'+str(j)+'.jpg')),rotated)

        
        
```
